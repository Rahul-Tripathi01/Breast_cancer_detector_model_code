```python
X=cancer_df.drop(["target"],axis=1)
```

```python
X.head()
```

```python
y=cancer_df["target"]
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=5)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train_sc=sc.fit_transform(X_train)
X_test_sc=sc.transform(X_test)
```

```python
from sklearn.metrics import confusion_matrix,classification_report,accuracy_score
```

```python
from sklearn.svm import SVC
svc_classifier= SVC()
svc_classifier.fit(X_train,y_train)
y_pred_scv =svc_classifier.predict(X_test)
accuracy_score(y_test,y_pred_scv)
```

```python
svc_classifier2= SVC()
svc_classifier2.fit(X_train_sc,y_train)
y_pred_scv_sc =svc_classifier2.predict(X_test_sc)
accuracy_score(y_test,y_pred_scv_sc)
```

```python
from sklearn.linear_model import LogisticRegression
lr1=LogisticRegression(random_state=51, penalty='l2')
lr1.fit(X_train,y_train)
y_pred=lr1.predict(X_test)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.linear_model import LogisticRegression
lr1=LogisticRegression(random_state=51, penalty='l2')
lr1.fit(X_train_sc,y_train)
y_pred=lr1.predict(X_test_sc)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier(n_neighbors=10,metric="minkowski",p=2)
knn.fit(X_train,y_train)
y_pred_knn=knn.predict(X_test)
accuracy_score(y_pred_knn,y_test)
```

```python
knn2=KNeighborsClassifier(n_neighbors=10,metric="minkowski",p=2)
knn2.fit(X_train_sc,y_train)
y_pred_knn2=knn2.predict(X_test_sc)
accuracy_score(y_test,y_pred_knn2)
```

```python
#naive basis classifier
from sklearn.naive_bayes import GaussianNB
nb_classifier= GaussianNB()
nb_classifier.fit(X_train,y_train)
y_pred_nb=nb_classifier.predict(X_test)
accuracy_score(y_test,y_pred_nb)
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dt=DecisionTreeClassifier()
dt.fit(X_train,y_train)
y_pred_dt=dt.predict(X_test)
accuracy_score(y_test,y_pred_dt)
```

```python
#decision tree classifier
from sklearn.tree import DecisionTreeClassifier
dt2=DecisionTreeClassifier()
dt2.fit(X_train_sc,y_train)
y_pred_dt_sc=dt2.predict(X_test_sc)
accuracy_score(y_test,y_pred_dt_sc)
```

```python
#random forest classifier
from sklearn.ensemble import RandomForestClassifier
rft=RandomForestClassifier()
rft.fit(X_train,y_train)
y_pred_rft=rft.predict(X_test)
accuracy_score(y_test,y_pred_rft)
```

```python
rft1=RandomForestClassifier()
rft1.fit(X_train_sc,y_train)
y_pred_rft1=rft1.predict(X_test_sc)
accuracy_score(y_test,y_pred_rft1)
```

```python
#XGboost classifier
from xgboost import XGBClassifier
xgb=XGBClassifier()
xgb.fit(X_train,y_train)
y_pred_xgb=xgb.predict(X_test)
accuracy_score(y_test,y_pred_xgb)
```

```python
# is it overfit?
```

```python
cm=confusion_matrix(y_test,y_pred_xgb)
plt.title('heatmap of the confusion matrix',fontsize=15)
sns.heatmap(cm,annot=True)
```

```python
print(classification_report(y_test,y_pred_xgb))
```

# Cross validation

```python
import numpy as np
import pandas as pd
from flask import Flask, request, render_template
import pickle

app = Flask(__name__)
model = pickle.load(open('model.pkl', 'rb'))

@app.route('/')
def home():
    return render_template('index.html')

@app.route('/predict',methods=['POST'])
def predict():
    input_features = [float(x) for x in request.form.values()]
    features_value = [np.array(input_features)]
    
    features_name = ['mean radius', 'mean texture', 'mean perimeter', 'mean area',
       'mean smoothness', 'mean compactness', 'mean concavity',
       'mean concave points', 'mean symmetry', 'mean fractal dimension',
       'radius error', 'texture error', 'perimeter error', 'area error',
       'smoothness error', 'compactness error', 'concavity error',
       'concave points error', 'symmetry error', 'fractal dimension error',
       'worst radius', 'worst texture', 'worst perimeter', 'worst area',
       'worst smoothness', 'worst compactness', 'worst concavity',
       'worst concave points', 'worst symmetry', 'worst fractal dimension']
    
    df = pd.DataFrame(features_value, columns=features_name)
    output = model.predict(df)
        
    if output == 0:
        res_val = "** breast cancer **"
    else:
        res_val = "no breast cancer"
        

    return render_template('index.html', prediction_text='Patient has {}'.format(res_val))

if __name__ == "__main__":
    app.run()
```
